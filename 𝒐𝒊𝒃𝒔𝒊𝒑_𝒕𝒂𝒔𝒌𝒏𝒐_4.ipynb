```python
#tokenizations starts from here in this way
#to show punctuations
import string
mess='Sample message ! Notice : it has punctuation.'
string.punctuation
```

```python
#removing punctuations marks
non_punc=[c for c in mess if c not in string.punctuation]
non_punc
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
non_punc=''.join(non_punc)
non_punc.split()
```

```python
#removing both punctuation and stop words
clean_mess=[word for word in non_punc.split() if word.lower() not
            in stopwords.words('english')]
clean_mess
```

```python
# a function that removes punctuation and stop words convert sentence into an array as text processing
def text_process(mess):
    nopunc=[char for char in mess if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return [word for word in nopunc.split() if word.lower not in stopwords.words('english')]
```

```python
data['sentence'].head(5).apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer=CountVectorizer(analyzer=text_process).fit(data['sentence'])
```

```python
print(len(bow_transformer.vocabulary_))
```

```python
mess4=data['sentence'][3]
```

```python
bow4=bow_transformer.transform([mess4])
```

```python
messages_bow=bow_transformer.transform(data['sentence'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data['sentence'],data['target'])
```

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([('bow',CountVectorizer(analyzer=text_process)),
                   ('tf_idf',TfidfTransformer()),
                   ('classifier',MultinomialNB())

])
```

```python
pipeline.fit(x_train,y_train)
```

```python
preds=pipeline.predict(y_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(preds,y_test))
```

```python
print('accuracy:',accuracy_score(preds,y_test))
```
